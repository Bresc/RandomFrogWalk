## Random walk Problema de la rana

```python
import numpy as np
def random_walk(n): #n is no. of steps
   x = 0
   xposition = [0] # origin
   for i in range (1,n): #from 1 to n
       step = np.random.uniform(0,1) #probability between 0 - 1
       if step < 0.5: # if step is less than 0.5 we move right    
           x += 1
       if step > 0.5: # if step is greater than 0.5 we move left  
           x += -1
       xposition.append(x) #add the move to vector xposition
   return xposition #Return the positions vector

Randwalk = random_walk(100000) #creating an object for the Randomwalk1D class and passing value of n
print(Randwalk) #prints the positions vector
```

```python
import pylab 
import random
import time
start_time = time.time()
n = 10000000 #n is the number of steps(increase in the value of n increses the compelxity of graph) 
x = np.zeros(n) # x and y are arrays which store the coordinates of the position 
y = np.zeros(n)
por = ''
tiempo = 0
direction=["NORTH","SOUTH","EAST","WEST"] # Assuming the four directions of movement.
for i in range(1, n): 
    step = random.choice(direction) #Randomly choosing the direction of movement.
    if step == "EAST": #updating the direction with respect to the direction of motion choosen.
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
        if x[i] == 250.0 and y[i] == 300.0 :
            por = str(i)
            print("--- %s seconds ---" % (time.time() - start_time))
    elif step == "WEST": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
        if x[i] == 250.0 and y[i] == 300.0 :
            por = str(i)
            print("--- %s seconds ---" % (time.time() - start_time))
    elif step == "NORTH": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
        if x[i] == 250.0 and y[i] == 300.0 :
            por = str(i)
            print("--- %s seconds ---" % (time.time() - start_time))
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
        if x[i] == 250.0 and y[i] == 300.0 :
            por = str(i)
            print("--- %s seconds ---" % (time.time() - start_time))
pylab.title("Random Walk 2-D")
pylab.plot(x, y) #plotting the walk.
pylab.show()
print("--- %s seconds ---" % (time.time() - start_time), "lo encontre en :" + por)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #for plotting the 3-D plot.
start_time = time.time()
print("--- 0.8606913089752197 seconds ---")
N =100000000
R = (np.random.rand(N)*6).astype("int") #Randomly intializing the steps
x = np.zeros(N) 
y = np.zeros(N)
z = np.zeros(N)
x[ R==0 ] = -1; x[ R==1 ] = 1 #assigning the axis for each variable to use
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x) #The cumsum() function is used to get cumulative sum over a DataFrame or Series axis i.e. it sums the steps across for eachaxis of the plane.
y = np.cumsum(y)
z = np.cumsum(z)
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.9) #alpha sets the darkness of the path.
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
print("--- %s seconds ---" % (time.time() - start_time), "lo encontre en :" + por)
```
